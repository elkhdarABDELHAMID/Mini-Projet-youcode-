```python
import pandas as pd
import matplotlib.pyplot as plt

# Lire les données de fichier excel
df = pd.read_excel(r'C:\Users\hamid\Desktop\Projet_youcode\data-sell4all.xlsx')

df.to_csv('dataset-sell4all.csv')
```

```python
# Afficher les 5 premieres lignes
print(df.head())
```

```python
# Afficher un résumé technique 
print(df.info())
```

```python
# Vérifier les noms des colonnes
print(df.columns)
```

```python
# Nettoyer la colonne ' Age '
df[' Age '] = pd.to_numeric(df[' Age '], errors='coerce')

# Nettoyer la colonne 'Dépenses des clients'
df['Dépenses des clients'] = pd.to_numeric(df['Dépenses des clients'], errors='coerce')

# Calculer la médiane et la moyenne de l'âge
age_median = df[' Age '].median()
age_mean = df[' Age '].mean()

# Calculer la médiane et la moyenne des dépenses des clients
spending_median = df['Dépenses des clients'].median()
spending_mean = df['Dépenses des clients'].mean()

print(f"Âge médian : {age_median}")
print(f"Âge moyen : {age_mean}")
print(f"Dépenses médianes : {spending_median}")
print(f"Dépenses moyennes : {spending_mean}")
```

print(f"Médiane de l'âge: {age_median}, Moyenne de l'âge: {age_mean}")
print(f"Médiane des dépenses: {spending_median}, Moyenne des dépenses: {spending_mean}")

```python
# Visualisation des données
df.groupby('Pays')['Dépenses des clients'].sum().plot(kind='bar')
plt.title('Dépenses des clients par pays')
plt.xlabel('Pays')
plt.ylabel('Dépenses des clients')
plt.show()
```

```python
# Nettoyage des données
df_cleaned = df[df['Dépenses des clients'] >= 10].drop_duplicates()
```

```python
# Enregistrement des données nettoyées
df_cleaned[['Pays', 'Age', 'Genre', 'Dépenses des clients']].to_csv('cleaned_dataset-sell4all.csv', index=False)
```
