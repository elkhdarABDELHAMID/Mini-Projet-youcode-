## 7 - Détection des anomalies (outliers) :

```python

# nettoyer les données de Montant_Rgl
df['Montant_Rgl'] = pd.to_numeric(df['Montant_Rgl'], errors='coerce')  # Convertir en numérique, gérer les erreurs

# Calculer les quartiles pour la colonne 'Montant_Rgl'
Q1 = df['Montant_Rgl'].quantile(0.25)
Q3 = df['Montant_Rgl'].quantile(0.75)
IQR = Q3 - Q1

# Définir les limites supérieures et inférieures pour identifier les anomalies
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identifier les transactions qui sont en dehors des limites
anomalies = df[(df['Montant_Rgl'] < lower_bound) | (df['Montant_Rgl'] > upper_bound)]

# Afficher les transactions identifiées comme anomalies
print("Transactions identifiées comme anomalies :")
print(anomalies)

```

```python
import matplotlib.pyplot as plt

# Tracer un histogramme pour visualiser la distribution des montants des transactions
plt.figure(figsize=(10, 6))
plt.hist(df['Montant_Rgl'].dropna(), bins=30, color='skyblue', edgecolor='black')  # Ignorer les NaN pour l'histogramme
plt.axvline(x=lower_bound, color='red', linestyle='dashed', linewidth=2, label='Limite Inférieure')
plt.axvline(x=upper_bound, color='green', linestyle='dashed', linewidth=2, label='Limite Supérieure')
plt.title("Histogramme des montants des transactions")
plt.xlabel("Montant_Rgl")
plt.ylabel("Fréquence")
plt.legend()
plt.show()
```

## 9- Analyse de la corrélation :

```python

# Assurez-vous que les colonnes 'Solde_CPP' et 'Montant_Rgl' existent dans votre DataFrame
if 'Solde_CPP' not in df.columns or 'Montant_Rgl' not in df.columns:
  print("Les colonnes 'Solde_CPP' ou 'Montant_Rgl' n'existent pas dans le DataFrame.")
else:
  # Grouper les données par client et calculer le solde moyen des cartes prépayées et les dépenses totales
  client_analysis = df.groupby('Bénéficiaire_CPP').agg({'Solde_CPP': 'mean', 'Montant_Rgl': 'sum'})

  # Calculer la corrélation entre le solde des cartes prépayées et les dépenses totales
  correlation = client_analysis['Solde_CPP'].corr(client_analysis['Montant_Rgl'])

  print("Corrélation entre le solde des cartes prépayées et les dépenses totales :")
  print(correlation)

  # Créer un scatter plot pour visualiser la relation
  plt.figure(figsize=(8, 6))
  plt.scatter(client_analysis['Solde_CPP'], client_analysis['Montant_Rgl'], alpha=0.5)
  plt.xlabel("Solde moyen des cartes prépayées")
  plt.ylabel("Dépenses totales")
  plt.title("Relation entre le solde des cartes prépayées et les dépenses des clients")
  plt.grid(True)
  plt.show()
```
